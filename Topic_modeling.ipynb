```python
import re

# Function to remove Arabic diacritics and other noise
def clean_text(text):
    # Regular expression to remove Arabic diacritics
    arabic_diacritics = re.compile(r'[\u0617-\u061A\u064B-\u0652]')

    # Remove diacritics
    text = re.sub(arabic_diacritics, '', text)

    # Remove other noise (special characters, punctuation)
    text = re.sub(r'[^\w\s]', '', text)  # Remove all non-alphanumeric characters except whitespace

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
cleaned_poems = [clean_text(poem) for poem in poems]

```
